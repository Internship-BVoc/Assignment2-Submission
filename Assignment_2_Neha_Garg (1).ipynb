**Problem 7**

***Are Divorced Individuals of higher age compared to Individuals that are Not-in-family ? Using mean as a measure , answer the above question & mention the respective age of the two groups***

```
p = df[(df['marital-status'] == ' Divorced') & (df['relationship'] != ' Not-in-family')]['age'].mean(0)
p
```

```
q = df[(df['relationship'] == ' Not-in-family') & (df['marital-status'] != ' Divorced')]['age'].mean(0)
q
```

```
if(p>q):
  print('True')
else:
  print('false')
```

**Conclusion- The age of divorced individuals are higher than the age of individuals that are not in family.**
