**Problem 7 :Are Divorced Individuals of higher age compared to Individuals that are Not-in-family ? Using mean as a measure , answer the above question & mention the respective age of the two groups.**

```
df_employee[ (df_employee['marital-status'] == ' Divorced')][['age']].mean()
```

```
df_employee[ (df_employee['relationship'] == ' Not-in-family')][['age']].mean()
```

Inferences: 
From above we come to know that the average age of Divorced Individuals is 42.0307 & the Individuals that are Not -in-family is 37.48 .Therefore ,the Divorced individuals has the higher age as compared to that of Not-in-family.
