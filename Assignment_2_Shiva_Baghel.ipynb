**2.Carry on a Univariate analysis on all the Variables/Fields. User bar plots for Categorical Variables & histogram or line plot for Continuous ones. Label the plot and axis clearly for intuitive understanding. Mention observations drawn from each Univariate plot**


```
a = df.age.hist()
a.set_xlabel('Age')
a.set_ylabel('Count of Employee')
a.set_title('Age v/s CoE')
```

**Observations**


1.   Maximum employee are of age (20 - 45)
2.   There are less employee b/w age range 60 or older



```
b = df.salary.hist(bins=[0,200000, 400000, 600000, 800000, 1050000],figsize=[15,4])
b.set_xlabel('Salary of Employee')
b.set_ylabel('Count of Employee')
b.set_title('Salary v/s CoE')
```

**Observations**


1.   Maximum Employee has salary below Rs. 2,00,000
2.   Very very few employee has salary above 4,00,000



```
b = df.hours_per_week.hist()
b.set_title('Working Hour v/s CoE')
b.set_xlabel('Employee working hour')
b.set_ylabel('Count of Employee')
```

**Observations**


1.   There are maximum employee worked for 30 - 40 hour/week (Outlier)
2.   About 100 employee works for 10 - 30 and 40-70 hour/week
3.   Very few employee work for 70 - 100 hours/week
