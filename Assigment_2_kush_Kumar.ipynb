**How many Males that have Marital status Married-civ-spouse are earning above 150,000 ?**

```
males=data_csv[(data_csv['Gender'] == ' Male') & (data_csv['marital-status'] == ' Married-civ-spouse') & (data_csv['salary']>150000)]['Gender'].count()
print(f"Males that have Marital status Married-civ-spouse are earning above 150,000 is {males}")
```

**Are Divorced Individuals of higher age compared to Individuals that are Not-in-family ? Using mean as a measure , answer the above question & mention the respective age of the two groups**

```
dev=data_csv[(data_csv['marital-status'] == ' Divorced') & (data_csv['relationship'] != ' Not-in-family')]['age'].mean(0)
dev1=data_csv[(data_csv['relationship'] == ' Not-in-family') & (data_csv['marital-status'] != ' Divorced')]['age'].mean(0)
print(dev)
print(dev1)
```
