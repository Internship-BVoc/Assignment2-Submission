### **6.How many Males that have Marital status Married-civ-spouse are earning above 150,000 ?**

```
s=df_emp[(df_emp['Gender'] == ' Male') & (df_emp['marital-status'] == ' Married-civ-spouse') & (df_emp['salary'] > 150000) ]
print(s)
print('Total  Number of Married civilian spouse  which earn more than 150000 are ',s.shape[0])
```

### **7.Are Divorced Individuals of higher age compared to Individuals that are Not-in-family ? Using mean as a measure , answer the above question & mention the respective age of the two groups.**

```
m1=df_emp[(df_emp['marital-status'] == ' Divorced') & (df_emp['relationship'] != ' Not-in-family')]['age'].mean()
m2=df_emp[(df_emp['relationship'] == ' Not-in-family') & (df_emp['marital-status'] != ' Divorced')]['age'].mean()
print('The mean age of individuals that are divorced :',m1)
print('The mean age of individuals that are not in family :',m2)
```

So,The age of  divorced individuals are higher  than the age of individuals that are not in family.
