```
6.How many Males that have Marital status ‘Married-civ-spouse’ are earning above 150,000 ?
```

```
nvg=df[(df["marital-status"] ==" Married-civ-spouse") & (df['Gender'] == " Male") & (df['salary'] > 150000)]['Gender'].count()
print("Married civilian =",nvg)
```

```
7.Are Divorced Individuals of higher age compared to Individuals that are Not-in-family ? Using mean as a measure , answer the above question & mention the respective age of the two groups
```

```
nv=df[(df['marital-status'] == ' Divorced') & (df['relationship'] != ' Not-in-family')]['age'].mean(0)
nv2=df[(df['relationship'] == ' Not-in-family') & (df['marital-status'] != ' Divorced')]['age'].mean(0)
print(nv)
print(nv2)

```
